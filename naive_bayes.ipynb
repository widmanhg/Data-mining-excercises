```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Example dataset
# Two features (Feature1, Feature2) and a target (Class)
data = {
    "Feature1": [1.1, 2.0, 1.3, 3.6, 1.8, 3.0, 3.9, 4.5, 5.2, 3.2],
    "Feature2": [2.2, 2.9, 1.5, 3.7, 2.5, 3.1, 4.8, 4.4, 5.1, 3.6],
    "Class": [0, 1, 0, 1, 0, 1, 1, 1, 1, 1],
}
```

```python
# Convert to a DataFrame
df = pd.DataFrame(data)

# Split dataset into features (X) and target (y)
X = df[["Feature1", "Feature2"]]
y = df["Class"]

df
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize Gaussian Naive Bayes classifier
gnb = GaussianNB()

# Fit the model
gnb.fit(X_train, y_train)

# Predict on test data
y_pred = gnb.predict(X_test)
```

```python
# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Example prediction
example = [[3.0, 2.0]]  # New data point
prediction = gnb.predict(example)
print("\nPrediction for example {}: Class {}".format(example, prediction[0]))
```
